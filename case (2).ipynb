```python
caminho = r'D:\Curso Python\Case\Dados_Case_IM.xlsx'
abas = pd.read_excel(caminho, sheet_name=None)
df_vendas = abas['Vendas por Rep']
df_visitas = abas['Visitação']
df_metas = abas['Critérios Prêmio']

```

```python
# 2. Automatização dos Dicionários de Regras
# 1. Definindo as metas
meta_A, meta_B, meta_C = 110000, 90000, 40000

# 2. Calculando o atingimento individual (%) 
df_vendas['Perc_A'] = df_vendas['Produto A'] / meta_A
df_vendas['Perc_B'] = df_vendas['Produto B'] / meta_B
df_vendas['Perc_C'] = df_vendas['Produto C'] / meta_C

# 3. Média Ponderada 

df_vendas['Performance_Final'] = (df_vendas['Perc_A'] * 0.5) + \
                                 (df_vendas['Perc_B'] * 0.3) + \
                                 (df_vendas['Perc_C'] * 0.2)


```

```python
print(df_vendas[['Produto A', 'Produto B', 'Produto C']].head())
```

```python
# 4. Calculando o Prêmio (Regra do Acelerador) [cite: 17, 28]
def aplicar_acelerador(perf):
    premio_base = 3000
    if perf >= 1.01: # Atingiu mais de 101% 
        # O acelerador começa em 125% e sobe conforme a performance 
        # Aqui usamos o multiplicador de 1.25 que descobrimos na tabela 
        return min(perf * 1.25, 1.50) * premio_base 
    return perf * premio_base

df_vendas['Premio_Final'] = df_vendas['Performance_Final'].apply(aplicar_acelerador)

# 5. Mostrando o resultado para conferir
print(df_vendas[['Representante', 'Performance_Final', 'Premio_Final']])
```

```python
# 1. Processando a aba de Visitas
df_visitas = abas['Visitação'] # Nome da aba no seu Excel
df_visitas.columns = df_visitas.columns.str.strip() # Limpeza de segurança

# Somando os 3 ciclos para ter o esforço total
cols_visitas = ['Visitas Ciclo 1', 'Visitas Ciclo 2', 'Visitas Ciclo 3']
df_visitas['Total_Visitas'] = df_visitas[cols_visitas].sum(axis=1)

# 2. Fazendo o MERGE 
# Unimos df_vendas com o df_visitas usando a coluna 'Representante'
df_consolidado = pd.merge(df_vendas, df_visitas[['Representante', 'Total_Visitas']], on='Representante')

# 3. Criando o KPI de Eficiência
# Soma de todos os produtos vendidos dividida pelo total de visitas
df_consolidado['Unidades_por_Visita'] = (df_consolidado['Produto A'] + 
                                         df_consolidado['Produto B'] + 
                                         df_consolidado['Produto C']) / df_consolidado['Total_Visitas']

# 4. Exportação Final para o Power BI
df_consolidado.to_excel('Base_Dashboard_Gilead.xlsx', index=False)
```
