### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

### Passo 4 - Calcular o indicador para 1 loja

```python
#definição de metas
meta_faturamento_dia = 1000
met_faturamento_ano = 1650000
meta_qtdprodutos_dia = 4
meta_qtdeprodutos_ano = 120
meta_ticketmedio_dia = 500
meta_ticketmedio_ano = 500
```

```python
import win32com.client as win32
import pathlib

# ==========================================
# DEFINIÇÕES INICIAIS (FORA DO LOOP)
# ==========================================

# 1. CRIANDO A LÓGICA DE CORES E ÍCONES:
# Definimos a função fora do 'for' para não criá-la repetidas vezes na memória
def avaliar_meta(valor, meta):
    if valor >= meta:
        return 'green', '▲'  # Verde com seta p/ cima se atingiu/superou a meta
    else:
        return 'red', '▼'    # Vermelho com seta p/ baixo se ficou abaixo da meta

# ==========================================
# INÍCIO DO LOOP PELAS LOJAS
# ==========================================

for loja in dicionario_lojas:
    # --------------------------------------
    # 1. CÁLCULO DOS INDICADORES
    # --------------------------------------
    
    # Aqui você pega o DataFrame da loja de dentro do dicionário
    vendas_loja = dicionario_lojas[loja]

    # CORREÇÃO: Use 'vendas_loja' (que é o DataFrame) para filtrar, não o dicionário
    vendas_loja_dia = vendas_loja.loc[vendas_loja['Data'] == dia_indicador, :]

    # Faturamento
    faturamento_ano = vendas_loja['Valor Final'].sum()
    faturamento_dia = vendas_loja_dia['Valor Final'].sum()

    # Diversidade de produtos
    qtd_produtos_ano = len(vendas_loja['Produto'].unique())
    # CORREÇÃO: Atenção ao nome da variável! Ajustado para 'qtde_produtos_dia' 
    # para bater com a chamada da função 'avaliar_meta' lá embaixo.
    qtd_produtos_dia = len(vendas_loja_dia['Produto'].unique())

    # Ticket medio
    valor_venda = vendas_loja.groupby('Código Venda')[['Valor Final']].sum()
    ticket_medio_ano = valor_venda['Valor Final'].mean()

    valor_venda_dia = vendas_loja_dia.groupby('Código Venda')[['Valor Final']].sum()
    ticket_medio_dia = valor_venda_dia['Valor Final'].mean()

    # --------------------------------------
    # 2. CONFIGURAÇÃO E ENVIO DO E-MAIL
    # --------------------------------------
    
    outlook = win32.gencache.EnsureDispatch('outlook.application')

    
    nome = emails.loc[emails['Loja'] == loja, 'Gerente'].values[0]
    mail = outlook.CreateItem(0)
    mail.To = emails.loc[emails['Loja'] == loja, 'E-mail'].values[0]
    mail.Subject = 'OnePage dia {}/{} - Loja {}'.format(dia_indicador.day, dia_indicador.month, loja)
    
    # Avaliando se bateu a meta para definir cor e ícone do dia
    cor_fat_dia, cenario_fat_dia = avaliar_meta(faturamento_dia, meta_faturamento_dia)
    cor_qtde_dia, cenario_qtde_dia = avaliar_meta(qtd_produtos_dia, meta_qtdprodutos_dia)
    cor_ticket_dia, cenario_ticket_dia = avaliar_meta(ticket_medio_dia, meta_ticketmedio_dia)
    
    # 2. MELHORANDO O VISUAL DO E-MAIL:
    # (Dica: Usei :,.2f para deixar os valores monetários com precisão de 2 casas decimais)
    corpo_email = f'''
    <div style="font-family: Calibri, Arial, sans-serif; color: #333333;">
        <p>Bom dia, <strong>{nome}</strong>!</p>
        <p>O resultado de ontem <strong>({dia_indicador.day}/{dia_indicador.month})</strong> da <strong>Loja {loja}</strong> foi:</p>
        
        <table style="border-collapse: collapse; min-width: 500px; text-align: center; margin-top: 20px;">
          <tr style="background-color: #004d99; color: white;">
            <th style="padding: 10px; border: 1px solid #dddddd;">Indicador</th>
            <th style="padding: 10px; border: 1px solid #dddddd;">Valor Dia</th>
            <th style="padding: 10px; border: 1px solid #dddddd;">Meta Dia</th>
            <th style="padding: 10px; border: 1px solid #dddddd;">Cenário Dia</th>
           </tr>
           
          <tr>
            <td style="padding: 10px; border: 1px solid #dddddd; text-align: left; background-color: #f9f9f9;">Faturamento</td>
            <td style="padding: 10px; border: 1px solid #dddddd; color: {cor_fat_dia}; font-weight: bold;">R$ {faturamento_dia:,.2f}</td>
            <td style="padding: 10px; border: 1px solid #dddddd;">R$ {meta_faturamento_dia:,.2f}</td>
            <td style="padding: 10px; border: 1px solid #dddddd; color: {cor_fat_dia}; font-weight: bold;">{cenario_fat_dia}</td>
          </tr>
          <tr>
            <td style="padding: 10px; border: 1px solid #dddddd; text-align: left;">Diversidade de Produtos</td>
            <td style="padding: 10px; border: 1px solid #dddddd; color: {cor_qtde_dia}; font-weight: bold;">{qtde_produtos_dia}</td>
            <td style="padding: 10px; border: 1px solid #dddddd;">{meta_qtdprodutos_dia}</td>
            <td style="padding: 10px; border: 1px solid #dddddd; color: {cor_qtde_dia}; font-weight: bold;">{cenario_qtde_dia}</td>
          </tr>
          <tr>
            <td style="padding: 10px; border: 1px solid #dddddd; text-align: left; background-color: #f9f9f9;">Ticket Médio</td>
            <td style="padding: 10px; border: 1px solid #dddddd; color: {cor_ticket_dia}; font-weight: bold;">R$ {ticket_medio_dia:,.2f}</td>
            <td style="padding: 10px; border: 1px solid #dddddd;">R$ {meta_ticketmedio_dia:,.2f}</td>
            <td style="padding: 10px; border: 1px solid #dddddd; color: {cor_ticket_dia}; font-weight: bold;">{cenario_ticket_dia}</td>
          </tr>
        </table>
        
        <p style="margin-top: 25px;">Segue em anexo a planilha com todos os dados para mais detalhes.</p>
        <p>Qualquer dúvida, estou à disposição.</p>
        <p>Att.,<br><strong>Juan</strong></p>
    </div>
    '''
    
    mail.HTMLBody = corpo_email 
    
    # Ajuste no nome do arquivo 
    arquivo_nome = f'{dia_indicador.month}_{dia_indicador.day}_{loja}.xlsx'
    
    # Atenção: Garanta que as variáveis "caminho_backup" e "dia_indicador" já estejam definidas acima no seu código original
    caminho_arquivo = pathlib.Path.cwd() / caminho_backup / loja / arquivo_nome
    
    # O SEGREDO DO SUCESSO: Caminho absoluto e .resolve()
    if caminho_arquivo.exists():
        caminho_absoluto = str(caminho_arquivo.resolve())
        print(f"Tentando anexar: {caminho_absoluto}")
        mail.Attachments.Add(caminho_absoluto)
        
        mail.Send()
        print(f"E-mail da loja {loja} enviado com sucesso, amor!")
    else:
        print(f"ERRO: Arquivo não encontrado em {caminho_arquivo}")



```
